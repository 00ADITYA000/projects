```python
import numpy as np
import pandas as pd
import datetime as dt
import datetime
```

```python
rw = pd.read_csv("data.csv",encoding='unicode_escape')
```

```python
rw.head()
```

```python
rw['Total'] = rw.Quantity * rw.UnitPrice
```

```python
rw.isna().sum()
```

```python
rw.head()
```

```python
rw.dropna(subset = ['Description','CustomerID'], inplace =True)
```

```python
rw
```

```python
now = dt.date(2011,12,30)
```

```python
rw['Date'] = pd.DatetimeIndex(rw['InvoiceDate']).date
```

```python
rw.head()
```

```python
rw_recency = rw.groupby('CustomerID',as_index=False)['Date'].max()
```

```python
rw_recency.columns
```

```python
rw_recency.head()
```

```python
rw_recency['Recency'] = rw_recency['Date'].apply(lambda x: (now-x).days)
```

```python
rw_recency.head()
```

```python
rw_frequency = rw.groupby('CustomerID',as_index = False)['InvoiceNo'].count()
```

```python
rw_frequency.columns=['CustomerID', 'Frequency']
```

```python
rw_frequency.head()
```

```python
rw_monetary = rw.groupby('CustomerID',as_index=False)['Total'].sum()
```

```python
rw_monetary.columns = ['CustomerID', 'Monetary']
```

```python
rw_monetary.head()
```

```python
RF = pd.merge(rw_recency,rw_frequency)
```

```python
RFM = pd.merge(RF,rw_monetary)
```

```python
RFM.head()
```

```python
quantile = RFM.quantile(q=[0.25,0.5,0.75])
quantile
```

```python
{'Recency':{0.25:37.0 , 0.50:71.0 , 0.75:164.0},
 'Frequency':{0.25:17.0 , 0.50:42.0 , 0.75:102.0},
 'Monetary':{0.25:293.3625 , 0.50:648.0750 , 0.75:1611.7250}}
```

```python
def Rscore(x,y,z):
    if x<=z[y][0.25]:
        return 1
    elif x<=z[y][0.50]:
        return 2
    elif x<=z[y][0.75]:
        return 3
    else:
        return 4
```

```python
def Fscore(x,y,z):
    if x<=z[y][0.25]:
        return 1
    elif x<=z[y][0.50]:
        return 2
    elif x<=z[y][0.75]:
        return 3
    else:
        return 4
```

```python
def Mscore(x,y,z):
    if x<=z[y][0.25]:
        return 1
    elif x<=z[y][0.50]:
        return 2
    elif x<=z[y][0.75]:
        return 3
    else:
        return 4
```

```python
RFM['R-score'] = RFM['Recency'].apply(Rscore, args = ('Recency',quantile))
RFM['F-score'] = RFM['Frequency'].apply(Fscore, args = ('Frequency',quantile))
RFM['M-score'] = RFM['Monetary'].apply(Mscore, args = ('Monetary',quantile))
```
